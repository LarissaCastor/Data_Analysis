```python
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive', force_remount= True)

df = pd.read_excel ('/content/drive/MyDrive/Dados econômicos CEPEA_20241210103934.xls' )
df = df.reset_index(drop=True)
df = df.set_index('Série')

queijo_mussarela = df.loc['preço queijo muçarela']
```

```python
#Foi necessário tratar a coluna 'Ano' convertendo para 'datetime objeto'
queijo_mussarela['Ano'] = pd.to_datetime(queijo_mussarela['Ano'], format='%Y').dt.year
queijo_mussarela.head(10)
```

```python
# gerar uma tabela com a média dos valores da coluna 'Brasil' agrupada para cada 'Ano'
```

```python
media_brasil_por_ano = queijo_mussarela.groupby('Ano')['Brasil'].mean()
media_brasil_por_ano.head(10)
```

```python
# Visualizar os dados em gráficos tornando-os mais didáticos e discutíveis
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(media_brasil_por_ano.index, media_brasil_por_ano.values)

plt.xlabel('Ano')
plt.ylabel('Média do Preço Brasil')
plt.title('Média do Preço do Queijo Mussarela no Brasil por Ano')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Realizar a mesma avaliação para o preço do leite. Pensando em correlacionar os valores.

leite_pasteurizado = df.loc['preço do leite pasteurizado']
leite_pasteurizado['Ano'] = pd.to_datetime(leite_pasteurizado['Ano'], format='%Y').dt.year
media_brasil_por_ano2 = leite_pasteurizado.groupby('Ano')['Brasil'].mean()
media_brasil_por_ano2.head(10)
```

```python
# Visualizar a média dos preços em um único gráfico visando avaliar uma linearidade entre os valores

plt.figure(figsize=(8, 5))
plt.bar(media_brasil_por_ano.index, media_brasil_por_ano.values, label='Queijo Mussarela')
plt.plot(media_brasil_por_ano2.index, media_brasil_por_ano2.values, marker='o', linestyle='-', color='red', label='Leite Pasteurizado')
plt.xlabel('Ano')
plt.ylabel('Média do Preço (R$)')
plt.title('Comparação da Média de Preços no Brasil')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Estudar a variação do preço entre os anos estudados

data = {
    'Ano': [2005, 2024],
    'Queijo Mussarela (R$)': [media_brasil_por_ano[2005] if 2005 in media_brasil_por_ano else None, media_brasil_por_ano[2024] if 2024 in media_brasil_por_ano else None],
    'Leite Pasteurizado (R$)': [media_brasil_por_ano2[2005] if 2005 in media_brasil_por_ano2 else None, media_brasil_por_ano2[2024] if 2024 in media_brasil_por_ano2 else None]
}

df_tabela = pd.DataFrame(data)
df_tabela
```

```python
queijo_variacao = ((media_brasil_por_ano[2024] - media_brasil_por_ano[2005]) / media_brasil_por_ano[2005]) * 100

leite_variacao = ((media_brasil_por_ano2[2024] - media_brasil_por_ano2[2005]) / media_brasil_por_ano2[2005]) * 100

print(f"Variação percentual do preço do Queijo Mussarela entre 2005 e 2024: {queijo_variacao:.2f}%")
print(f"Variação percentual do preço do Leite Pasteurizado entre 2005 e 2024: {leite_variacao:.2f}%")
```
