```
PIB = df[['Período','AGROPECUÁRIA', 'INDÚSTRIA', 'SERVIÇOS']]
```

```
PIB.head(3)
```

```
PIB.info()
```

```
PIB['AGROPECUÁRIA'] = PIB['AGROPECUÁRIA'].apply(lambda x: str(x).replace(",","."))
PIB['INDÚSTRIA'] = PIB['INDÚSTRIA'].apply(lambda x: str(x).replace(",","."))
PIB['SERVIÇOS'] = PIB['SERVIÇOS'].apply(lambda x: str(x).replace(",","."))
PIB.head(3)
```

```
PIB['AGROPECUÁRIA'] = PIB['AGROPECUÁRIA'].astype('float64')
PIB['AGROPECUÁRIA']
```

```
PIB['INDÚSTRIA'] = PIB['INDÚSTRIA'].astype('float64')
PIB['INDÚSTRIA']
```

```
PIB['SERVIÇOS'] = PIB['SERVIÇOS'].astype('float64')
PIB['SERVIÇOS']
```

```
PIB.head(3)
```

```
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```
dados = PIB[['Período','AGROPECUÁRIA', 'INDÚSTRIA', 'SERVIÇOS']]
dados = dados.melt('Período', var_name='Reporte', value_name='Taxa_Trimestral')
dados['Taxa_Trimestral'] = dados['Taxa_Trimestral'].astype('float64')
display(dados)
```

```
np_array = df.to_numpy()
per = df['Período'].to_numpy()
per
```

```
plt.figure(figsize=(20,5))
sns.color_palette("viridis", as_cmap=True)
sns.barplot(data=dados, x='Período', y='Taxa_Trimestral', hue='Reporte', dodge=False)
plt.xticks(rotation=90)
```

```
#Para visualização Gráfico de linhas
plt.figure(figsize=(20,5))
sns.lineplot(data=dados, x='Período', y='Taxa_Trimestral', hue='Reporte')
plt.xticks(rotation=90)
```
