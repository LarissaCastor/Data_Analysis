```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
```

```python
df = pd.read_excel ('/content/drive/MyDrive/finance_date/dados_Dol_Fut_nov22.xlsx' , header=0)
df
```

```python
#Renomeando colunas
df = df.rename(
    columns={
        "Fechamento": "DOLfut"
           }
)
df.head()
```

```python
#excluindo colunas
df.drop('Abertura', axis = 1, inplace=True)
df.drop('Máxima', axis = 1, inplace=True)
df.drop('Mínima', axis = 1, inplace=True)
df.drop('Amplitude', axis = 1, inplace=True)
```

```python
df.head()
```

```python
df2 = pd.read_excel ('/content/drive/MyDrive/finance_date/dados_DI1Fut.xlsx' , header=0)
df2.head()
```

```python
#Renomeando colunas
df2 = df2.rename(
    columns={
        "Fechamento": "DIfut"
           }
)
```

```python
#excluindo colunas
df2.drop('Abertura', axis = 1, inplace=True)
df2.drop('Máxima', axis = 1, inplace=True)
df2.drop('Mínima', axis = 1, inplace=True)
df2.drop('Amplitude', axis = 1, inplace=True)
```

```python
df2.head()
```

```python
df3 = pd.read_excel ('/content/drive/MyDrive/finance_date/dados_IND_Fut_nov22.xlsx' , header=0)
df3.head()
```

```python
#Renomeando colunas
df3 = df3.rename(
    columns={
        "Fechamento": "INDfut"
           }
)
```

```python
#excluindo colunas
df3.drop('Abertura', axis = 1, inplace=True)
df3.drop('Máxima', axis = 1, inplace=True)
df3.drop('Mínima', axis = 1, inplace=True)
df3.drop('Amplitude', axis = 1, inplace=True)
```

```python
df3.head()
```

```python
#Indexando coluna 
df = df.set_index("Data")
df.head()
```

```python
#Indexando coluna 
df2 = df2.set_index("Data")
df2.head()
```

```python
#Indexando coluna 
df3 = df3.set_index("Data")
df3.head()
```

```python
from numpy.ma.core import append
#unindo (concatenando) tabelas
table1 = pd.concat([df, df2, df3], axis=1)
table1
```

```python
import seaborn as sns
```

```python
sns.heatmap(table1.corr(),
            vmin=-1, vmax=1, annot=True, cmap='BrBG')
```

```python
table2 = table1.reset_index()
table2
```

```python
table2.set_index('Data', append=True)
```

```python
#Selecionando período de tempo específico devido a acontecimentos políticos -> eleições
table2 = table2.loc[4897:4918]
table2
```

```python
sns.heatmap(table2.corr(),
            vmin=-1, vmax=1, annot=True, cmap='BrBG')
```
